```python
import keyboard
import pygetwindow as gw

# Dictionary to hold input for each window
window_inputs = {}
current_window = None

def on_key_event(event):
    global current_window
    
    # Check for 'F1' key to exit the program
    if event.name == 'f1':
        print("Exiting program...")
        keyboard.unhook_all()  # Stop listening to all keyboard events
        return
    
    # Get the active window's title
    active_window = gw.getActiveWindow()
    window_title = active_window.title if active_window else "Unknown Window"
    
    # If the active window has changed, initialize a new input for it
    if window_title != current_window:
        current_window = window_title
        if current_window not in window_inputs:
            window_inputs[current_window] = ""

    # Handle key inputs
    if event.name == 'space':
        window_inputs[current_window] += " "  # Add a space for space key
    elif event.name == 'backspace':
        window_inputs[current_window] = window_inputs[current_window][:-1]  # Remove the last character
    elif len(event.name) == 1:  # Only append single character keys
        window_inputs[current_window] += event.name

    print(f"Current Input in '{current_window}': {window_inputs[current_window]}")

# Start listening for global key presses
keyboard.on_press(on_key_event)

print("Press 'F1' to stop the program...")

# Wait until 'F1' is pressed to exit the program
keyboard.wait('f1')
```
